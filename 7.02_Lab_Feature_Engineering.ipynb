# Activity

Create a function that takes a dataframe as an input and a percentual threshold (default value = 0.25) and returns a list of columns with null values greater than the specified threshold

```python
# Define the function here
def columns_above_threshold(df: pd.DataFrame, threshold: float=0.25) -> list:
    
    # Calculate the percentage of null values for each column
    null_percentages=pd.DataFrame(df.isnull().sum()/len(df)).reset_index()
    null_percentages.columns = ["column","percentage"]

    # Filter columns based on the threshold
    above_threshold_columns = null_percentages[null_percentages['percentage'] > threshold]

    # Extract column names as a list
    above_threshold_columns_list = above_threshold_columns['column'].values.tolist()

    return above_threshold_columns_list
```

```python
# Use your function here
columns_above_threshold_list = columns_above_threshold(data, threshold=0.25)
display(columns_above_threshold_list)
```

```python
#create a df with:
#the panda series of categorical columns and see their frequencies, value_counts, 
#use reset_index, to have the state not as a key, but as a label, new column

df = pd.DataFrame(categorical['STATE'].value_counts()).reset_index()


df.columns = ['state', 'count']
#see what are the states that have an amount of customers smaller than 2500 and pick column state, cast it as a list
other_states = list(df[df['count']<2500]['state'])

#_____
#define a function
#for every single state, value in a column (other_states), pick the state (value) and see if it is added on the column
#otherwhise return "other" 

# other_states = ['CA','AR','MD']
def clean_state(x):
    if x in other_states:
        return 'other'
    else:
        return x

#apply the function:

# clean_states(x,other_states) for x, other_states in zip(categorical['STATE'].values,other_states)
categorical['STATE'] = list(map(clean_state, categorical['STATE'])) #use the function into every single element of this column and cast the vlaues with a list

#categorical['STATE'] = categorical['STATE'].apply(clean_state)
```
